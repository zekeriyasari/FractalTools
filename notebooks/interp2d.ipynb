# 2D Fractal Interpolation 

In this notebook, we consider the interpolation of the data points given as 
$$
P = \{ (x_i, y_i, z_i)~|~ x_i, y_i, z_i \in \mathbb{R}, \; i = 1, 2, \ldots, N \}
$$
where $N$ is the number of points. The points $(x_i, y_i) \in \Omega \subset \mathbb{R}^2$. The aim is to find a continous function, also called as interpolant $f: \Omega \mapsto \mathbb{R}$ such the interpolation points are satisfied, i.e. $f(x_i , y_i) = z_i, \; i  = 1, 2, \ldots, N$.

The type of IFS has the form 
$$
w_n(x, y, z) =
\begin{bmatrix}
a_{11, n} & a_{12,n} & 0 \\ 
a_{21, n} & a_{22,n} & 0 \\ 
a_{31, n} & a_{32,n} & a_{33,n} \\ 
\end{bmatrix}
\begin{bmatrix}
x \\ 
y \\ 
z
\end{bmatrix}
+ 
\begin{bmatrix}
b_{1, n} \\ 
b_{2,n } \\ 
b_{3, n}
\end{bmatrix}
$$

Given the free variables $a_{33, n}, \; n = 1, 2, \ldots, N$, the remaining unknown coefficients in the transformations can be calculated using boundary conditions. 
$$
\begin{aligned}
w_n(x_0, y_0, z_0) &= (x_{n - 1}, y_{n - 1}, z_{n - 1}) \\ 
w_n(x_N, y_N, z_N) &= (x_{n}, y_{n}, z_{n})
\end{aligned}
$$
Hence, from the boundary conditions, we have 
$$
\begin{aligned}
    a_{11, n} x_0 + a_{12, n} y_0 + b_{1, n} &= x_{n - 1} \\
     
    a_{11, n} x_N + a_{12, n} y_N + b_{1, n} &= x_{n}
\end{aligned}
$$
$$
\begin{aligned}
    a_{21, n} x_0 + a_{22, n} y_0 + b_{1, n} &= x_{n - 1} \\
     
    a_{11, n} x_N + a_{12, n} y_N + b_{1, n} &= x_{n}
\end{aligned}
$$
